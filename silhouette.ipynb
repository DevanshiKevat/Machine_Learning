```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
# Generate sample data with 3 clusters
X, _ = make_blobs(n_samples=100, centers=3, n_features=2, random_state=42)
```

```python
# Plot the generated data
plt.scatter(X[:, 0], X[:, 1], c='red', marker='.', edgecolor='black')
plt.title("Generated Sample Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
# Initialize KMeans clustering algorithm
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
# Fit KMeans to the data
kmeans.fit(X)
```

```python
# Predict the cluster labels
labels = kmeans.labels_
labels
```

```python
# Calculate silhouette score
silhouette_avg = silhouette_score(X, labels)
print("The average silhouette_score is :", silhouette_avg)
```

```python
conversion_rate = 75  # 1$ = 75 Rs

# Function to clean and convert price values to INR
def convert_to_inr(price):
    if isinstance(price, str) and price.startswith('$'):
        price_in_usd = float(price[1:])
        return price_in_usd * conversion_rate
    elif isinstance(price, float):
        return price * conversion_rate
    else:
        return price

# Apply the function to 'Price' column
df[' Discounts '] = df[' Discounts '].apply(convert_to_inr)
df[' Units Sold '] = df[' Units Sold '].apply(convert_to_inr)
df[' Manufacturing Price '] = df[' Manufacturing Price '].apply(convert_to_inr)
df[' Sale Price '] = df[' Discounts '].apply(convert_to_inr)
df[' Gross Sales '] = df[' Gross Sales '].apply(convert_to_inr)
df[' Sales '] = df[' Sales '].apply(convert_to_inr)
df[' COGS '] = df[' COGS '].apply(convert_to_inr)
df['  Sale '] = df['  Sale '].apply(convert_to_inr)
df[' Profit '] = df[' Profit '].apply(convert_to_inr)

# Convert the 'Price' column to float
df[' Discounts '] = pd.to_numeric(df[' Discounts '], errors='coerce')

# Check the data types
df.dtypes
```
